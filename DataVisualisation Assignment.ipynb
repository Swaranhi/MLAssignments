# Data Visualisation Assignment

<b>In this assignment students have to transform iris data into 3 dimensions and plot a 3d chart with transformed dimensions and colour each data point with specific class.<b> 


```python
import numpy as np 
import matplotlib.pyplot as plt 
from mpl_toolkits.mplot3d import Axes3D 
from sklearn import decomposition 
from sklearn import datasets 
import seaborn as sns
from sklearn.decomposition import PCA
import pandas as pd
```

```python
iris_data=sns.load_dataset('iris')
```

```python
iris_data.head()
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(iris_data.iloc[:,0:4])

```

```python
pca_result
```

```python
df=pd.DataFrame(pca_result,columns=['p1','p2','p3'])
```

```python
df
```

```python
fig = plt.figure(figsize = (15,15))
ax = plt.axes(projection='3d')
species = ['setosa', 'versicolor', 'virginica']
colors = ['r', 'g', 'b']
for spc,colour in zip(species,colors):
    i=iris_data['species']==spc
    ax.scatter(
    df.loc[i,"p1"], 
    df.loc[i,"p2"], 
    df.loc[i,"p3"], 
    s=50,c=colour
)
ax.set_xlabel('pca-one')
ax.set_ylabel('pca-two')
ax.set_zlabel('pca-three')
ax.legend(species)
plt.show()
```
